```python
import pandas as pd
```

```python
visitation_df = pd.read_csv("./Resources/Annual_Visitation.csv")
visitation_df.head()
```

```python
parks_df = pd.read_csv("./Resources/parks.csv")
parks_df.head()
```

```python
species_df = pd.read_csv("./Resources/species.csv")

species_df.head()
```

```python
for col in species_df.columns:
    print(col)
```

```python
join_df = species_df.merge(parks_df, left_index=True, right_index=True)
join_df= join_df.drop(['Park Name_y'], axis=1)
join_df.head()
```

```python
join_df = join_df.merge(visitation_df, left_index=True, right_index=True)
join_df= join_df.drop(['Park Name_x'], axis=1)
join_df.head()
```

```python
for col in join_df.columns:
    print(col)
```

```python
join_df.to_csv('joined_df.csv', index=False)
```
